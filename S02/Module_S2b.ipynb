### Introduction
Recall that a Normal distribution with mean $ \mu $ and standard deviation $ \sigma $ is denoted by $ N(\mu , \sigma)$  
  
The distribution describes some **_population_** and is centered at $ \mu $ (the **_population mean_**) and the spread of the distribution is proportional to $ \sigma $ (the **_population standard deviation_**).


When we take a **sample** from that population, the _sample mean_ is denoted by $\overline{x}$.  

The sample mean is often used as an _estimate_ of the true population mean $\mu$.  
  
Likewise, the _sample standard deviation_ is denoted by $s$ and is used as an estimate of $\sigma$.

### Sampling Distributions 

A **_sampling distribution_** is a distribution of a sample statistic, such as $\overline{x}$ or $S$.  In a **_sampling distribution of the mean_**, we can generate many samples and look at how the sample means ($\overline{x}$'s) from all the samples are distributed.

```r
all_means <- c()                              #Empty vector to store all the sample means
for (count in 1:100){
    sample <- rnorm(10,mean=mu, sd=sigma)     #Generate a sample (size n=10)
    all_means[count] <- mean(sample)          #Save the mean of this sample in my list
}

head(all_means)
summary(all_means)
cat("standard deviation of sampling distribution:", sd(all_means))
hist(all_means)
```

## Central Limit Theorem
#### For sufficiently large values of $n$, when samples of size $n$ are drawn from ANY population distribution with mean $\mu$ and standard deviation $\sigma$, the theoretical distribution of the **_Sampling Distribution of Means_** is:
## <center> $N(\mu, \frac{\sigma}{\sqrt{n}})$ </center>

####  For sufficiently large values of $n$, this relationship holds **_regardless of the shape of the original population distribution_**. 

### The standard deviation of a sampling distribution (in this case, $\frac{\sigma}{\sqrt{n}}$) is called the _Standard Error_ (SE).
